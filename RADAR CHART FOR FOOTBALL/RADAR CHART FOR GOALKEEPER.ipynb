```python
def createRadar(player, data):
    Attributes = ["GA","PSxG","PSxG/SoT","PSxG+/-","Stp%","#OPA/90","GA90","SoTA","Saves","Save%","CS%"]
    
    data += data [:1]
    
    angles = [n / 11* 2 * pi for n in range(11)]
    angles += angles [:1]
    
    ax = plt.subplot(111, polar=True)

    plt.xticks(angles[:-1],Attributes)
    ax.plot(angles,data)
    ax.fill(angles, data, 'blue', alpha=0.1)

    ax.set_title(player)
    plt.show()
    
```

```python
createRadar("Alisson",[-1.377946,-1.381789,-1.291544,-1.168495,0.258240,1.677580,-1.259376,-1.346129,-1.372619,1.430977,1.712716])
```
